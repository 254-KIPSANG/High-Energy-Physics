```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define constants
k = 1.380649e-23 # Boltzmann constant (J/K)
T = 300 # Temperature (K)
e = 1.602176634e-19 # Electron charge (C)
h = 6.62607015e-34 # Planck constant (J s)
m = 9.10938356e-31 # Electron mass (kg)
Ef = 0.3 # Fermi energy (eV)

```

```python
# Define energy range
E = np.linspace(-0.5,0.5,1001) # Energy range in eV

```

```python
# Calculate Fermi-Dirac distribution
F = 1./(1+np.exp((E-Ef)/(k*T/e)))

```

```python
# Plot Fermi-Dirac distribution
plt.figure(1)
plt.plot(E,F,'k',linewidth=2)
plt.title('Fermi-Dirac Distribution')
plt.xlabel('Energy (eV)')
plt.ylabel('Occupation Probability')
plt.ylim([0,1])
plt.grid(True)

```

```python
# Define band parameters
Ec = 0.4 # Conduction band energy (eV)
Ev = -0.1 # Valence band energy (eV)
Eg = Ec - Ev # Band gap energy (eV)

```

```python
# Calculate density of states
Nc = 2*((2*np.pi*m*k*T)/(h**2))**(3/2)*np.exp(-(Ec-Ef)/(k*T)) # Effective density of states in conduction band
Nv = 2*((2*np.pi*m*k*T)/(h**2))**(3/2)*np.exp(-(Ef-Ev)/(k*T)) # Effective density of states in valence band

```

```python
# Calculate carrier concentrations
n = Nc*np.exp((Ef-Ec)/(k*T)) # Electron concentration
p = Nv*np.exp((Ev-Ef)/(k*T)) # Hole concentration

```

```python
# Plot energy band diagram
plt.figure(2)
plt.plot([0,Eg],[Ec,Ec],'r-',linewidth=2) # Conduction band
plt.plot([0,Eg],[Ev,Ev],'b-',linewidth=2) # Valence band
plt.plot([0,Eg],[Ef,Ef],'k--',linewidth=1) # Fermi level
plt.title('Energy Band Diagram')
plt.xlabel('Energy (eV)')
plt.ylabel('Energy Level (eV)')
plt.ylim([Ev-0.1,Ec+0.1])
plt.grid(True)
plt.legend(['Conduction Band','Valence Band','Fermi Level'])

plt.show()
```
